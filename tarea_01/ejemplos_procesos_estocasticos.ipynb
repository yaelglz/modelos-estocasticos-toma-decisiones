## 1. Gráficas de Modelos Estocásticos

Como ejemplo de un proceso estocástico, en este ejercicio se realiza la descarga de datos del rendimiento de los **Cetes a 28 días** para generar una curva de plazos para los valores de los Cetes de un año entre el **2003 y el 2026**.


### Fuente de Datos

Los datos de los Cetes en México se obtienen de la siguiente página:

[https://acortar.link/jG1jxn](https://acortar.link/jG1jxn)


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("rendimiento_cetes.xlsx", skiprows=18, header=None)
df.columns = ["fecha", "rendimiento"]

df.head()
```

### Creación de la gráfica

A continuación, se elaborará una gráfica del rendimiento de los **Cetes a 28 días** usando los datos cargados en el DataFrame `df`. En esta visualización, el eje horizontal representará la **fecha** y el eje vertical el **rendimiento**, con el objetivo de observar su comportamiento a lo largo del tiempo.


```python
plt.figure(figsize=(10,5))
plt.plot(df["fecha"], df["rendimiento"])

plt.title("Rendimiento de CETES a 28 días")
plt.xlabel("Fecha")
plt.ylabel("Rendimiento (%)")

plt.xticks(rotation=45)
plt.show()
```

## 2. Ejemplo de Proceso de Markov

Suponga que va a medir un proceso estocástico de Markov con estado inicial $X_0 = 0$ tiene probabilidad de aumento de una unidad de $1/2$ y probabilidad de disminuir en $1/2$.

Calcule los primeros 10 pasos de la caminata aleatoria: $x_1(\omega), x_2(\omega), x_3(\omega), \dots, x_{10}(\omega)$.

Haga una gráfica donde muestra cómo evoluciona el proceso de Markov.


```python
import numpy as np

# Configuración inicial
x0 = 0
n_pasos = 10
probabilidad = 0.5

# Generación de los pasos
pasos = np.random.choice([1, -1], size=n_pasos, p=[probabilidad, 1 - probabilidad])

# El proceso es la suma acumulada de los pasos
x_markov = np.concatenate(([x0], np.cumsum(pasos)))

# Mostrar los valores calculados
print(f"Valores del proceso de Markov (x0 a x10):")
for i, val in enumerate(x_markov):
    print(f"x{i}(\u03C9) = {val}")
```

### Evolución del Proceso de Markov

A continuación se muestra la gráfica de la caminata aleatoria.


```python
plt.figure(figsize=(10, 5))
plt.step(range(n_pasos + 1), x_markov, where='post', marker='o', color='green', label='Caminata Aleatoria')
plt.axhline(0, color='black', linestyle='--', linewidth=0.8)

plt.title("Evolución del Proceso de Markov (Caminata Aleatoria)")
plt.xlabel("Pasos (n)")
plt.ylabel("Estado (Xn)")
plt.xticks(range(n_pasos + 1))
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
