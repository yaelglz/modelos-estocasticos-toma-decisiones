# Distribución de Poisson
En esta tarea se resuelven problemas relacionados con la distribución de Poisson, definida por la función de probabilidad:
$$f(x) = P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!}, \quad k \in \{0, 1, 2, ...\}$$
Donde $\lambda$ es el promedio de eventos por unidad de tiempo o espacio.


```python
from scipy.stats import poisson
```

### 1. Problema Veterinaria
La veterinaria de Jorge recibe un promedio de $\lambda = 5$ pacientes por día. El número de pacientes sigue una distribución de Poisson.

Calcule:
1. La probabilidad de que lleguen 3 pacientes en un día.
2. La probabilidad de que lleguen 5 pacientes en un día.
3. La probabilidad de que no llegue ningún paciente.


```python
lam_vet = 5

# 1.1 La probabilidad de que lleguen 3 pacientes en un día.
p1_1 = poisson.pmf(3, lam_vet)
print(f"1.1 P(X = 3) = {p1_1:.4f}")

# 1.2 La probabilidad de que lleguen 5 pacientes en un día.
p1_2 = poisson.pmf(5, lam_vet)
print(f"1.2 P(X = 5) = {p1_2:.4f}")

# 1.3 La probabilidad de que no llegue ningún paciente.
p1_3 = poisson.pmf(0, lam_vet)
print(f"1.3 P(X = 0) = {p1_3:.4f}")
```

### 2. Problema de Grúas
El número de solicitudes de ayuda recibidas por un servicio de grúas sigue una distribución de Poisson con un promedio de 4 llamadas por hora ($\lambda = 4$ por hora).

Calcule:
1. La probabilidad de que exactamente 10 solicitudes sean recibidas durante un período particular de 2 horas.
2. La probabilidad de que se reciban menos de 5 llamadas en una hora.
3. La probabilidad de que se reciban entre 3 y 5 llamadas en una hora.


```python
lam_gruas_hora = 4

# 2.1 Probabilidad de exactamente 10 solicitudes en 2 horas.
# Para 2 horas, lambda se ajusta: lam = 4 * 2 = 8
lam_2horas = 8
p2_1 = poisson.pmf(10, lam_2horas)
print(f"2.1 P(X = 10 en 2h) = {p2_1:.4f}")

# 2.2 Probabilidad de que se reciban menos de 5 llamadas (en 1 hora).
# P(X < 5) = P(X <= 4)
p2_2 = poisson.cdf(4, lam_gruas_hora)
print(f"2.2 P(X < 5) = {p2_2:.4f}")

# 2.3 Probabilidad de que se reciban entre 3 y 5 llamadas (en 1 hora).
# P(3 <= X <= 5) = P(X = 3) + P(X = 4) + P(X = 5)
p2_3 = poisson.pmf(3, lam_gruas_hora) + poisson.pmf(4, lam_gruas_hora) + poisson.pmf(5, lam_gruas_hora)
print(f"2.3 P(3 <= X <= 5) = {p2_3:.4f}")
```
